```python
import pandas as pd
```

```python
# read in csv for apple data
apple = pd.read_csv('AAPL.csv')
# read in csv for AMD data
amd = pd.read_csv('AMD.csv')
# read in csv for Google data
google = pd.read_csv('GOOG.csv')
# read in csv for Intel data
intel = pd.read_csv('INTC.csv')
# read in csv for Microsoft data
microsoft = pd.read_csv('MSFT.csv')
# read in csv for Nvidia data
nvidia = pd.read_csv('NVDA.csv')
# read in csv for Tesla data
tesla = pd.read_csv('TSLA.csv')
```

```python
# grouping datasets together for addition of new columns
dfs = [apple, amd, google, intel, microsoft, nvidia, tesla]
```

```python
# applying 50 and 200 day moving averages to all dataframes
for df in dfs:
    df ['MA50'] = df.Close.rolling(50).mean()
    df ['MA200'] = df.Close.rolling(200).mean()
```

```python
# making column for previous days price at close
for df in dfs:
    df ['Previous Day Close Price'] = df.Close.shift(1)
```

```python
amd.head()
```

```python
# new column for change in price calculated by selected days price minus previous days close price
for df in dfs:
    df ['Change in Price'] = df['Close'] - df['Previous Day Close Price']
```

```python
amd.head()
```

```python
# create new column for percentage of change in price
for df in dfs:
    df ['Percent Change in Price'] = df.Close.pct_change()
```

```python
amd.head()
```

```python
# create new column for previous days volume
for df in dfs:
    df ['Previous Day Volume'] = df.Volume.shift(1)
```

```python
# create new co
for df in dfs:
    df ['Change in Volume'] = df['Volume'] - df['Previous Day Volume']
```

```python
for df in dfs:
    df ['Percent Change in Volume'] = df.Volume.pct_change()
```
